## Init

```python
%matplotlib inline
import random
import numpy as np
import pandas as pd
import seaborn as sns
from pprint import pprint
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
model_input = pd.read_csv("data/dataset_1.csv")
model_input.rename(columns={'nmin_90' : 'leach'}, inplace = True)
```

## Heatmap

```python
# Selecting only  numerical variables 
numerical_data = model_input.select_dtypes(include=['float64', 'int64'])

# Correlation matrix
correlation_matrix = numerical_data.corr()

# Plotting the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt = ".2f", cmap='vlag')
plt.title('Heatmap of Correlations')
plt.show()
```

## Histograms

## Transforms

```python
dfs = {}

# Separate columns
crop_column = model_input[['crop']]
doy_column = model_input[['doy']]

# Nmin conversion to binary
threshold = model_input['leach'].mean()
nmin_column = (model_input['leach'] > threshold).astype(int)

# Transform DOY into two dimensions using sine and cosine
doy_column['doy_sin'] = np.sin(2 * np.pi * doy_column['doy'] / 365)
doy_column['doy_cos'] = np.cos(2 * np.pi * doy_column['doy'] / 365)

# One-hot encode the 'crop' column
encoder = OneHotEncoder(sparse=False)
crop_encoded = encoder.fit_transform(crop_column)
crop_df = pd.DataFrame(crop_encoded, columns=encoder.get_feature_names_out(['crop']))

# Drop 'crop' and 'doy' columns and the specified 'field' from the original DataFrame
fields = model_input.drop(['crop', 'doy', 'leach'], axis=1)

# Scale the remaining numerical features
scaler = StandardScaler(with_mean=False)
scaled_fields = scaler.fit_transform(fields)
scaled_fields_df = pd.DataFrame(scaled_fields, columns=fields.columns)
```

```python
#split into sets to be analyzed together
yvar = 'leach'

# Concatenate the transformed 'crop', 'doy', and scaled numerical features
transformed_df = pd.concat([doy_column[['doy_sin', 'doy_cos']], crop_df, scaled_fields_df, nmin_column], axis=1)

# drops bands not in 4band spectrum and indices from them
fourband = transformed_df.drop(['band_1', 'band_3', 'band_5', 'band_7', 'NDRE', 'MCARI', 'LCCI'], axis = 1)

# drops rows with NA present (only occurs in 4band rows for extra bands)
NAmask  = transformed_df.apply(lambda x: x.notna().all(), axis=1)
eightband = transformed_df[NAmask]

# doesn't add crop columns and drops bands not in 4band spectrum and indices from them 
nocrop = pd.concat([doy_column[['doy_sin', 'doy_cos']], scaled_fields_df, nmin_column], axis=1)
nocrop_four = nocrop.drop(['band_1', 'band_3', 'band_5', 'band_7', 'NDRE', 'MCARI', 'LCCI'], axis = 1)

# doesn't add  crop columns and drops rows with NA present (only occurs in 4band rows for extra bands))
nocrop_eight = nocrop[NAmask]
```

## Transformed Heatmap

```python
# Selecting only  numerical variables 
numerical_data = transformed_df.select_dtypes(include=['float64', 'int64'])

# Correlation matrix
correlation_matrix = numerical_data.corr()

# Plotting the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, fmt = ".2f", cmap='vlag')
plt.title('Heatmap of Correlations')
plt.show()
```

## PCA

```python
#PCA

dfs = [
    [fourband, 'fourband'], 
    [eightband, 'eightband'],
    [nocrop_four, 'nocrop four'],
    [nocrop_eight, 'nocrop eight']
    ]

pcas = {}

for data, name in dfs:
    x = data.drop(yvar, axis=1)
    y = data[[yvar]]
    cn = data.columns #column names
    n = data.shape[0] #nrows
    results = {}
    explained_var_ratios = {}
    # Loop through the desired number of components
    explained_var_ratios = {}
    for n_components in range(2, 4):
        pca = PCA(n_components=n_components)
        pca_fit = pca.fit_transform(x)
        explained_var_ratio = sum(pca.explained_variance_ratio_) * 100
        explained_var_ratios[n_components] = explained_var_ratio
    pcas[name] = explained_var_ratios
    print(name,"(n =", n,")", ":",pcas[name])
```

## Feature Importance

```python
for data, name in dfs:
    x = data.drop(yvar, axis=1)
    y = data[[yvar]]
    cn = x.columns

    #set model 
    model=RandomForestClassifier(random_state=42, oob_score = True)
    model.fit(x, y.values.ravel());
    order = np.argsort(model.feature_importances_)
    oob_score = model.oob_score_
    
    #plot feature importances
    plt.figure(figsize=(15, 5))
    plt.bar(x=np.array(cn)[order],
            height=model.feature_importances_[order])
    plt.gca().set_ylabel('Importance', labelpad=10)
    plt.gca().set_xlabel('Feature', labelpad=10)
    plt.xticks(rotation='vertical')
    plt.text(0.02, 0.95, f'OOB Score: {oob_score:.2f}', transform=plt.gca().transAxes)
    plt.tight_layout()
    plt.title(name)
```
